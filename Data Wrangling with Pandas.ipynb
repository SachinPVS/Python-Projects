# Stakeholders are interested in obtaining the first and last names, street address, city, state, and information on any duplicate values in the customer personal details dataset.

```python
import pandas as pd
```

```python
customer_data = pd.read_csv('data/Customer_info.csv')
customer_data.head()
```

```python
customer_data = pd.read_csv('data/Customer_info.csv', header=None)
customer_data.head()
```

```python
customer_data.tail()
```

```python
customer_data.drop([2,3], axis=1, inplace=True)
customer_data.head()
```

```python
column_names = ["Full_name", "Address"]
customer_data.columns = column_names
```

```python
customer_data[['First_name', 'Last_name']] = customer_data.Full_name.str.split(expand=True)
customer_data.head()
```

```python
customer_data.drop('Full_name', axis=1, inplace=True)
customer_data.head()
```

```python
customer_data[['Street_address', 'City', 'State']] = customer_data.Address.str.split(pat = ", ", expand=True)
customer_data.head()
```

```python
customer_data.drop('Address', axis = 1, inplace=True)
customer_data.head()
```
