### For the year 2018 and 2019, import the credit sales data. Give the regions' minimum, maximum, average and overall sales figures.

```python
import pandas as pd
```

```python
credit_card_sales_demographics_2018 = pd.read_csv('data/credit_sales_2018.csv')
credit_card_sales_demographics_2019 = pd.read_csv('data/credit_sales_2019.csv')
```

```python
credit_card_sales_demographics_2018.head()
```

```python
credit_card_sales_demographics_2019.head()
```

```python
credit_card_sales_demographics_2018['year'] = 2018
credit_card_sales_demographics_2018.head()
```

```python
credit_card_sales_demographics_2019['year'] = 2019
credit_card_sales_demographics_2019.head()
```

```python
# Merging the datasets
credit_card_sales_demographics = pd.concat([credit_card_sales_demographics_2018, credit_card_sales_demographics_2019])
credit_card_sales_demographics.head()
```

```python
credit_card_sales_demographics = credit_card_sales_demographics.melt(id_vars=["year", "Region"], var_name=["demographic"],
                                                                     value_name="sales ($) in K")
credit_card_sales_demographics.head()
```

```python
credit_card_sales_demographics['gender'] = credit_card_sales_demographics.demographic.str[0]
credit_card_sales_demographics['age_group'] = credit_card_sales_demographics.demographic.str[1:]
```

```python
credit_card_sales_demographics.head()
```

```python
credit_card_sales_demographics.drop('demographic', axis=1, inplace=True)
```

```python
credit_card_sales_demographics.tail()
```

```python
credit_card_sales_demographics.sort_values(by = ['year', 'Region', 'gender'], inplace=True)
credit_card_sales_demographics.head()
```

```python
credit_card_sales_demographics.tail()
```

```python
# Reorder the columns
credit_card_sales_demographics = credit_card_sales_demographics[['Region', 'age_group', 'gender', 'year', 'sales ($) in K']]
credit_card_sales_demographics.head()
```

```python
credit_card_sales_demographics.reset_index(drop=True, inplace=True)
credit_card_sales_demographics.head()
```

```python
# Format the data two decimal places
pd.options.display.float_format = '{:,.2f}'.format
```

```python
pd.pivot_table(credit_card_sales_demographics, index = ['year', 'Region', 'gender', 'age_group'],  
               aggfunc = ['sum', 'min', 'max', 'mean'], margins = True, margins_name='Total')
```
